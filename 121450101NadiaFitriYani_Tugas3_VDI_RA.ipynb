```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# Membaca file Excel
df = pd.read_excel('lamteng2023.xlsx')

df
```

1. Disini saya menggunaka

```python
# Histogram: Distribusi Produksi Padi
plt.figure(figsize=(8, 6))
sns.histplot(df['hasil produksi'], bins=5, kde=True, color='purple')
plt.title('Distribusi Produksi Padi')
plt.xlabel('Produksi Padi (ton)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(df['luas tanam'], df['hasil produksi'])
plt.xlabel('Luas Tanam')
plt.ylabel('Hasil Produksi')
plt.title('Hubungan Luas Tanam dengan Hasil Produksi')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['luas panen'], df['hasil produksi'])
plt.xlabel('Luas Panen')
plt.ylabel('Hasil Produksi')
plt.title('Hubungan Luas Panen dengan Hasil Produksi')
plt.grid(True)
plt.show()
```

```python
# Plot bar
plt.figure(figsize=(12, 6))

# Plot bar untuk Luas Tanam
plt.bar(df['bulan'], df['luas tanam'], width=0.3, label='Luas Tanam', align='center')

# Plot bar untuk Luas Panen
plt.bar(df['bulan'], df['luas panen'], width=0.3, label='Luas Panen', align='edge')

# Plot bar untuk Hasil Produksi
plt.bar(df['bulan'], df['hasil produksi'], width=-0.3, label='Hasil Produksi', align='edge')

# Menambahkan judul dan label
plt.title('Perbandingan Luas Tanam, Luas Panen, dan Hasil Produksi Per Bulan', fontsize=14)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Luas dan Hasil', fontsize=12)

# Rotasi label sumbu x
plt.xticks(rotation=45)

# Menampilkan legenda
plt.legend()

# Menampilkan visualisasi
plt.tight_layout()
plt.show()
```

Hasil Visualisasi:
Garis Tren memberikan pandangan tentang perubahan bulan ke bulan secara halus untuk tiap variabel.
Bar Plot menyoroti perbandingan antar variabel dalam satu bulan dengan jelas, terutama untuk analisis distribusi.

```python
# Heatmap: Korelasi antar variabel
plt.figure(figsize=(8, 6))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi antar Variabel')
plt.show()
```
